```python
#data scraping using beautiful soup 

import pandas as pd
import numpy as np
import requests 

from bs4 import BeautifulSoup 

 

req = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M") 

soup = BeautifulSoup(req.content,'lxml') 

table = soup.find_all('table')[0]  

df = pd.read_html(str(table)) 

 

neighborhood=pd.DataFrame(df[0])

neighborhood.head()
```

```python
# Filter all rows for which the Borough is not assigned
df_filtered = neighborhood[neighborhood['Borough'] != 'Not assigned'].reset_index(drop=True)
  
# Print the new dataframe 
df_filtered.head(10)
```

```python
# Filter all rows for which the neighbourhood is not assigned


df_filtered['Neighbourhood'].replace('Not Assigned','Borough')
df_filtered.head(10)
```
